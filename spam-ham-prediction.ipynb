```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re

%matplotlib inline
```

```python
data=pd.read_csv("/kaggle/input/sms-spam-collection-dataset/spam.csv", encoding='ISO-8859-1')
```

```python
data.head()
```

```python
data=data.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
data.head(2)
```

```python
new_col_name={"v1":"Label","v2":"sms"}
for old_name,new_name in new_col_name.items():
    data.rename(columns={old_name:new_name},inplace=True)
    
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=data,x="Label")
plt.show()
```

```python
data.shape
```

```python
data.drop_duplicates()
data.shape
```

```python
data['Label']=data['Label'].map({"spam":1,"ham":0})
```

```python
data.head(2)
```

```python
x=data.iloc[:,-1]
y=data.iloc[:,0]
```

**Text Preprocessing:**

```python
x
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
lem=WordNetLemmatizer()
preprocessed_sms=[]
for text in x:
    text = re.sub('^[a-zA-Z]', " ", text)
    text = text.lower()
    text = re.sub(r'\d+', " ", text)
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    words = nltk.word_tokenize(text)
    # Join the words back into a string
    preprocessed_text = ' '.join([(word) for word in words if word not in set(stopwords.words("english"))])
    preprocessed_sms.append(preprocessed_text)
```

```python
preprocessed_sms[:3]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer()

x=cv.fit_transform(preprocessed_sms).toarray()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
x_test[:5]
```

**Prediction Model : Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
nb_model=MultinomialNB()
```

```python
nb_model.fit(x_train,y_train)
```

```python
y_pred=nb_model.predict(x_test)
```

**Model Validation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print('\n')
print(accuracy_score(y_test,y_pred))
print('\n')
print(classification_report(y_test,y_pred))
```
